```python
import geopandas as gp
```

Crop region shapefile from research collaborators...

```python
crop_regions = gp.read_file('/g/data/xv83/unseen-projects/outputs/wheatbelt/shapefiles/crops_SA_WA.shp')
```

```python
crop_regions = crop_regions.drop([1])
```

```python
crop_regions
```

```python
crop_regions.plot()
```

States and territories from the Australian Bureau of Statistics...

```python
states = gp.read_file('/g/data/ia39/aus-ref-clim-data-nci/shapefiles/data/aus_states_territories/aus_states_territories.shp')
```

```python
wa = states[states['STE_NAME21'] == 'Western Australia']
```

```python
wa.plot()
```

```python
sa = states[states['STE_NAME21'] == 'South Australia']
```

```python
sa.plot()
```

Create new cropping region shapefiles for WA and SA separately...

```python
wa_crop_regions = crop_regions.clip(wa)
```

```python
wa_crop_regions.plot()
```

```python
wa_crop_regions.to_file('/g/data/xv83/unseen-projects/outputs/wheatbelt/shapefiles/crops_WA.shp')
```

```python
sa_crop_regions = crop_regions.clip(sa)
```

```python
sa_crop_regions.plot()
```

```python
sa_crop_regions.to_file('/g/data/xv83/unseen-projects/outputs/wheatbelt/shapefiles/crops_SA.shp')
```
